```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../assets/COVID-19 Cases(02-10-2025).csv")
df
```

```python
df.describe()
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='mixed')
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
```

```python
df
```

```python
maharashtra_data = df[df['Region']== "Maharashtra"]
```

```python
maharashtra_data
```

```python
maharashtra_data_year_2021 = maharashtra_data[maharashtra_data['Year']==2021]
```

```python
maharashtra_data_year_2021
```

```python
plt.plot(maharashtra_data_year_2021['Month'],maharashtra_data_year_2021['Death'])
```

```python
maharashtra_data_year_2021_month_10 = maharashtra_data_year_2021[maharashtra_data_year_2021['Month']==10]
```

```python
maharashtra_data_year_2021_month_10
```

```python
plt.plot(maharashtra_data_year_2021_month_10['Day'],maharashtra_data_year_2021_month_10['Death'])
```

```python
maharashtra_data_year_2021_month_10.sort_values(by='Day', inplace=True)
```

```python
maharashtra_data_year_2021_month_10
```
