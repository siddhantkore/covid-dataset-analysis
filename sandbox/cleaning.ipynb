```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../assets/COVID-19 Cases(02-10-2025).csv")
df
```

```python
df.describe()
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='mixed')
df['Year'] =df['Date'].dt.year
df['Month'] =df['Date'].dt.month
df['Day'] =df['Date'].dt.day
df
```

```python
df.plot(x='Year', y='Death', kind='scatter')
```

Here is problem in dataset these are years such as 2017, 2018, 2019 which are not relevant to covid-19. So we will delete all rows with year<2020 and reset index.

```python
df = df[df['Year'] >= 2020]
df.reset_index(drop=True, inplace=True)
```

```python
df.plot(x='Year', y='Death', kind='scatter')
```

```python
df.plot(x='Month', y='Death', kind='scatter')
```

```python
maharashtra = df[df['Region'] == 'Maharashtra']
maharashtra
```

```python
maharashtra.plot(x='Month', y='Death', kind='line')
```

```python
maharashtra_2020 = maharashtra[maharashtra['Year'] == 2020]
maharashtra_2020
```

```python
maharashtra_2020.plot(x='Month', y='Death', kind='scatter')
```

```python
maharashtra_2020_grouped = maharashtra_2020.groupby(['Year','Month'])[['Confirmed Cases', 'Active Cases', 'Cured/Discharged', 'Death']].sum()
maharashtra_2020_grouped
```

```python
maharashtra_2020.columns
```

```python
maharashtra_2020_grouped.reset_index(inplace=True)
```

```python
maharashtra_2020_grouped.plot(x='Month', y='Death')
```
